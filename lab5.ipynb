#3. Алгоритмічне завдання (Bubble Sort + підрахунок перестановок)

```python
def bubble_sort_with_swaps(arr):
    if not isinstance(arr, list):
        raise TypeError("Потрібно ввести список.")
    swaps = 0
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swaps += 1
    return arr, swaps
```

#4. Тестові приклади

```python
print(bubble_sort_with_swaps([5, 3, 8, 4, 2]))   # ([2, 3, 4, 5, 8], кількість перестановок)
print(bubble_sort_with_swaps([1, 2, 3]))         # ([1, 2, 3], 0)
print(bubble_sort_with_swaps([10, -1, 0]))       # ([-1, 0, 10], кількість перестановок)
```

#5. Контроль введення

```python
try:
    print(bubble_sort_with_swaps("123"))
except Exception as e:
    print("Помилка:", e)
```

#6. Графічна візуалізація (matplotlib)

```python
import matplotlib.pyplot as plt

examples = [[5,3,8,4,2], [1,2,3], [10,-1,0]]
labels = ["Приклад 1", "Приклад 2", "Приклад 3"]
swaps = []

for ex in examples:
    _, count = bubble_sort_with_swaps(ex.copy())
    swaps.append(count)

plt.figure(figsize=(8,4))
plt.bar(labels, swaps, color="red")
plt.title("Кількість перестановок у Bubble Sort")
plt.ylabel("Перестановки")
plt.show()
```
